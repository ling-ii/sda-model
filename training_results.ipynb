```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = 'training/batch_0'
files = [os.path.join(path, file) for file in os.listdir(path)]
losses = [np.fromfile(file) for file in files]
epochs = np.arange(500)
```

```python
def plot(x, y, lab, ax):
    if len(x) != len(y):
        print(f"Cannot plot {lab}")
        return
    ax.plot(x, y, label=lab, linewidth=1)
```

```python
files.sort()
files
```

```python
fig, axs = plt.subplots(1, 1, figsize=(10,4))

for i, loss in enumerate(losses):
    if i > 3:
        continue
    plot(epochs, loss, str(i), axs)

fig.legend()
```
